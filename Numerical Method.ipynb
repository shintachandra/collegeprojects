```python
#akan diimport semua modul yang diperlukan untuk menurunkan fungsi, ekspresi matematik, membuat dataframe, dll
from tabulate import tabulate
import sympy
from numpy import sin,cos,tan,log,sqrt,exp,nan,pi
import numpy as np
from tabulate import tabulate
import math
import pandas as pd

print('Program ini akan menemukan akar persamaan f(x) = 0 secara numerik.') #kalimat penjelas

formula = input('Masukkan fungsi f(x) yang masih mengandung variabel t: ') #inputan fungsi yang masih mengandung variabel t disimpan ke dalam variabel formula
NPM = eval((input('Masukkan NPM Anda: '))) #inputan memasukkan NPM disimpan ke dalam variabel NPM

#dilakukan kondisi untuk mengecek apakah npm yang diinput ganjil atau genap
if NPM % 2 == 0:
    t = 0 #npm genap
    print('Maka, nilai t = 0')
else:
    t = 1 #npm ganjil
    print('Maka, nilai t = 1')
print(" ") #untuk line spacing
print('Akan diinput interval [a,b], boleh bergantung t') #kalimat untuk menujukkan step selanjutnya

low_bound = eval(input("Masukkan batas bawah interval (a): ")) #inputan batas bawah disimpan ke dalam variabel low_bound
up_bound = eval(input("Masukkan batas atas interval (b): ")) #inputan batas atas disimpan ke dalam variabel up_bound
print('Interval yang digunakan adalah ','[',low_bound,",",up_bound,']') #kalimat untuk melihat batas atau dan bawah
print('Akan digunakan tebakan awal = b yaitu',up_bound) #kalimat penjelas
print("")
tole = eval(input('Masukkan batas toleransi: ')) #inputan nilai toleransi disimpan ke dalam variabel tole
iterasi = eval(input('Masukkan batas banyaknya iterasi: ')) #inputan batas iterasi disimpan ke dalam variabel iterasi

#METODE BISECTION
Bisec_List=[] #untuk menyimpan nilai-nilai dari bisection ke dalam list bernama Bisec_List

def f(x,t): #fungsi untuk formula dengan 2 parameter (x,t)
    return eval(formula)

def Bisection(low_bound, up_bound, tole): #Fungsi untuk menjalankan bisection
#ALGORITMA BISECTION
    i =1 #inisiasi nilai awal i sama dengan 1
    if f(low_bound,t)*f(up_bound,t)<0: #syarat nilai batas atas dan bawah
        p0=low_bound
        p=(low_bound+up_bound)/2 #mencari titik p
        if f(p,t)==0:
            return p
        elif f(p,t)*f(low_bound,t)>0:
            low_bound=p
        elif f(p,t)*f(low_bound,t)<0:
            up_bound=p
            
        abs_error=abs(p0-p)
        p0=p
    
        while abs_error > tole and i <= iterasi: #membatasi iterasi sampai 50x
            p=(low_bound + up_bound)/2
            
            if f(p,t)==0:
                break
            elif f(p,t)*f(low_bound,t)>0: #syarat ketika hasil kali nilai fungsinya positif
                low_bound = p #memasukkan nilai p ke dalam var low_bound
            elif f(p,t)*f(low_bound,t)<0:#syarat ketika hasil kali nilai fungsinya negatif
                up_bound = p #memasukkan nilai p ke dalam var up_bound
                
            abs_error=abs(p0-p)
            p0=p
            Bisec_List.append(p) #mendata seluruh nilai p yang telah diproses ke dalam list Bisec_List
            i += 1 #menambahkan nilai i agar iterasi berjalan
        return p #mereturn nilai p

akar_bisection=Bisection(low_bound,up_bound,tole)

#METODE FIXED POINT
FP_list=[] #nilai akar dari metode fixed point disimpan ke dalam list bernama FP_list

#ALGORITMA FIXED POINT
def g(x,t): #fungsi untuk x-(f(x,t))
    return x-f(x,t)

def FixedPoint(up_bound,tole): #fungsi fixed point
    i = 1 #inisiasi nilai awal i =1
    p = g(up_bound,t) #insiasi nilai p
    abs_error = abs(p-up_bound)
    p0 = p #memasukkan nilai p ke p0
    
    while abs_error > tole and i <= iterasi: #membatasi iterasi sampai 50x
        i += 1 #menambah nilai i agar loop tetap berjalan
        p = g(up_bound,t) #inisiasi nilai p
        abs_error = abs(p-up_bound)
        up_bound = p
        
        FP_list.append(p) #mendata seluruh nilai p (akar fixed point) ke dalam list FP_list
        
    return p

fixed_point = FixedPoint(up_bound, tole)


#METODE NEWTON BIASA
Newton_list=[] #Membuat tampungan list untuk nilai dari metode newton

#ALGORITMA NEWTON BIASA
def f(x,t):
    return eval(formula)

x = sympy.symbols("x") #untuk menghitung turunan terhadap x

fp_string = str(sympy.diff(formula, x))

def fp(x,t):
    return eval(fp_string)

def Newton(up_bound, tole): #fungsi untuk newton
    i= 1 #inisiasi nilai awal i = 1
    p = up_bound - f(up_bound,t)/fp(up_bound,t)
    abs_error = abs(p-up_bound)
    up_bound = p
    
    while abs_error > tole and i<= iterasi: #membatasi iterasi sebanyak 50x
        try:
            p = up_bound - f(up_bound,t)/fp(up_bound,t)
        except ZeroDivisionError:
            return "Metode gagal, pilih tebakan lain"
        
        abs_error = abs(p-up_bound)
        up_bound = p
        i += 1 #menambahkan nilai i agar looping berjalan
       
        Newton_list.append(p) #mendata seluruh nilai p ke dalam list Newton_list

    return p
hasil = Newton(up_bound, tole)

#METODE FINITE DIFFERENCE
FD_list = [] #membuat tampungan untuk nilai dari finite different

#ALGORITMA FINITE DIFFERENCE
def f(x,t):
    return eval(formula)

def df(x, h=10**(-12)): #membuat fungsi df dengan parameter x dan nilai h = 10^-12
    return (f(x+h,t)-f(x,t))/h

def Finite_Difference(up_bound,tole): #fungsi finite diff
    i =1 #inisiasi nilai awal i=1
    p = up_bound - f(up_bound,t)/df(up_bound,t)
    abs_error = abs(p-up_bound)
    up_bound = p
    
    while abs_error > tole and i <= iterasi:#membatasi iterasi sampai 50x
        p = up_bound - f(up_bound,t)/df(up_bound,t)
        abs_error = abs(p-up_bound)
        up_bound = p
        FD_list.append(p) #untuk medata semua nilai p (nilai akar dengan metode Finite diff) ke dalam list FD_list
    
        i+= 1 #menambahkan nilai i agar looping tetap berjalan
  
    return p

akar_fd = Finite_Difference(up_bound,tole)

#METODE SECANT
Secant_list=[] #membuat tampungan untuk nilai akar dari metode secant
#ALGORITMA METODE SECANT
def f(x,t):
    return eval(formula)

def Secant(low_bound,up_bound,tole): #fungsi untuk menjalankan metode secant
    p = up_bound - (f(up_bound,t)*(up_bound-low_bound))/(f(up_bound,t)-f(low_bound,t))
    abs_error = abs(p-up_bound)
    low_bound = up_bound
    up_bound = p
    i = 1 #inisiasi nilai i = 1
    
    while abs_error > tole and i <= iterasi:
        p = up_bound - (f(up_bound,t)*(up_bound-low_bound))/(f(up_bound,t)-f(low_bound,t))
        abs_error = abs(p-up_bound)
        low_bound = up_bound
        up_bound = p
        
        Secant_list.append(p) #mendata seluruh nilai p ke dalam list Secant_list
       
        i += 1 #menambahkan nilai i agar loopong tetap berjalan

    return p

secant = Secant(low_bound,up_bound, tole)


#Metode Regula Falsi
Regula_Falsi = [] #tampungan untuk nilai akar metode regula falsi
#ALGORITMA REGULA FALSI
def f(x,t):
    return eval(formula)

def RegulaFalsi(low_bound, up_bound, tole): #fungsi untuk menjalankan metode Regula Falsi
    i = 1 #inisiasi i = 1
    if f(low_bound,t) * f(up_bound,t) < 0:
        p0 = low_bound
        p = (low_bound+up_bound)/2
        if f(p,t) == 0:
            return p
        elif f(p,t)*f(low_bound,t) > 0:
            low_bound = p
        elif f(p,t)*f(low_bound,t) < 0:
            up_bound = p
            
        abs_error = abs(p0-p)
        p0 = p
        
    
        while abs_error > tole and i <= iterasi: #membatasi iterasi hingga 50x
            p = up_bound - ((f(up_bound,t)*(up_bound-low_bound))/(f(up_bound,t)-f(low_bound,t)))
            if f(p,t) == 0:
                break
            elif f(p,t)*f(low_bound,t) > 0:
                low_bound = p
            elif f(p,t)*f(low_bound,t) < 0:
                up_bound = p
            abs_error = abs(p0-p)
            p0 = p
            
            
            Regula_Falsi.append(p) #mendata semua nilai p ke dalam var list Regula_Falsi
            
            i = i + 1 #menambahkan nilai i agar looping berjalan

        return p
    
Regula_falsi = RegulaFalsi(low_bound, up_bound, tole)


#STEFFENSEN
Steffensen_list=[] #tampungan untuk nilai akar dari metode steffensen
#ALGORITMA STEFFENSEN
def g(x):
    return x-f(x,t)

def Steffensen(up_bound, tole): #fungsi untuk menjalankan metode steffensen
    list_p = [up_bound] #list untuk batas atas
    abs_error = tole + 1
    
   
    i = 1 
    while abs_error >= tole and i <= iterasi:
        if i % 3 == 0: 
            pn_3 = list_p[i - 3] # p_(n-3)
            pn_2 = list_p[i - 2] # p_(n-2)
            pn_1 = list_p[i - 1] # p_(n-1)
            Delta1 = pn_2 - pn_3
            Delta2 = pn_1 - 2 * pn_2 + pn_3
            pn = pn_3 - (Delta1)**2 / Delta2
        else: 
            pn_1 = list_p[i - 1]
            pn = g(pn_1)
            
        list_p.append(pn)
        abs_error = abs( pn - pn_1 )
        
       
        Steffensen_list.append(pn) #menambahkan nilai pn (akar) ke dalam var list Steffensen_list
        
        i += 1 #menambahkan i agar looping tetap berjalan
        

    # return bukan hanya p, tetapi juga banyaknya iterasi
    return pn, iterasi


p_steffensen, i_steffensen = Steffensen(up_bound, tole)

#MENYAMAKAN PANJANG LIST DARI MASING-MASING LIST metode di atas 

#Mencari list mana yang paling panjang dari semua metode di atas, kemudian disimpan ke dalam variabel K
K = max(len(Bisec_List), len(FP_list), len(Newton_list), len(FD_list), len(Secant_list), len(Regula_Falsi)) 

#Selanjutnya kita ingin mengisi sisa list yang kosong dengan nilai kosong agar panjang tetap sama walau iterasi setiap metodenya berbeda
#M adalah variabel untuk menyimpan panjang list dari metode2 di atas
#dilakukan iterasi sebanyak selisih dari nilai terpanjang list dengan jumlah list masing2 metode
#lalu ke dalam selisih tersebut dimasukkan tanda V
M = len(Bisec_List)
for i in range(K-M):
    Bisec_List.append("V")
        
M = len(FP_list)
for i in range(K-M):
    FP_list.append("V")
        
M = len(Newton_list)
for i in range(K-M):
    Newton_list.append("V")
        
M = len(FD_list)
for i in range(K-M):
    FD_list.append("V")
        
M = len(Secant_list)
for i in range(K-M):
    Secant_list.append("V")
        
M = len(Regula_Falsi)
for i in range(K-M):
    Regula_Falsi.append("V")
        
M = len(Steffensen_list)
for i in range(K-M):
    Steffensen_list.append("V")
    
#MEMBUAT TABEL BESAR DENGAN METODE ZIP
#membuat variabel ALL untuk menyimpan gabungan dari list2 metode di atas
ALL = zip(Bisec_List, FP_list, Newton_list, FD_list, Secant_list, Regula_Falsi, Steffensen_list)
df = pd.DataFrame(ALL, columns=['Bisec_List','FP_list','Newton_list', 'FD_list', 'Secant_list','Regula_Falsi','Steffensen_list'])
#membuat var judul yang berisi list nama-nama kolom di tabel nanti
judul = ["i","Bisection", "Fixed Point", "Newton", "Finite Difference", "Secant", "Regula Falsi", "Steffensen"]
#Menyamakan panjang list
#Membuat output tabel besar
print(tabulate(df, headers = judul, tablefmt="pretty"))
```
