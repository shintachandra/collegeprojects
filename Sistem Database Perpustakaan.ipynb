# Pembuatan SQL

```python
# Buat koneksi ke database SQLite
conn = sqlite3.connect('perpustakaan.db')
cursor = conn.cursor()
cursor.execute('''DROP TABLE IF EXISTS buku''')
cursor.execute('''DROP TABLE IF EXISTS buku_per_copy''')
cursor.execute('''DROP TABLE IF EXISTS anggota''')
cursor.execute('''DROP TABLE IF EXISTS penjaga_perpustakaan''')
cursor.execute('''DROP TABLE IF EXISTS peminjaman''')
cursor.execute('''DROP TABLE IF EXISTS klasifikasi''')
cursor.execute('''DROP TABLE IF EXISTS penerbit''')
cursor.execute('''DROP TABLE IF EXISTS pengarang''')

# Buat tabel buku
cursor.execute('''CREATE TABLE buku (
                 ID_Buku TEXT PRIMARY KEY,
                 judul TEXT,
                 pengarang TEXT,
                 penerbit TEXT,
                 Tahun_terbit INTEGER,
                 klasifikasi TEXT,
                 FOREIGN KEY (pengarang) REFERENCES pengarang(id_pengarang),
                 FOREIGN KEY (penerbit) REFERENCES penerbit(id_penerbit),
                 FOREIGN KEY (klasifikasi) REFERENCES klasifikasi(id_klasifikasi)
                 )''')

# Buat tabel bukucopy
cursor.execute('''CREATE TABLE buku_per_copy (
                 id_buku_per_copy TEXT PRIMARY KEY,
                 Id_buku TEXT,
                 tersedia INTEGER,
                 FOREIGN KEY (Id_buku) REFERENCES buku(ID_Buku)
                 )''')

# Buat tabel anggota
cursor.execute('''CREATE TABLE anggota (
                 id_anggota TEXT PRIMARY KEY,
                 nama_anggota TEXT,
                 nik TEXT,
                 ttl TEXT,
                 alamat TEXT,
                 nomor_telepon TEXT)''')

# Buat tabel penjaga perpustakaan
cursor.execute('''CREATE TABLE penjaga_perpustakaan (
                 id_penjaga TEXT PRIMARY KEY,
                 nama_penjaga TEXT,
                 jadwal_kerja TEXT,
                 jam_kerja TEXT)''')

# Buat tabel peminjaman
cursor.execute('''CREATE TABLE peminjaman (
                 id_peminjaman TEXT PRIMARY KEY,
                 id_buku_per_copy TEXT,
                 id_anggota TEXT,
                 id_penjaga TEXT,
                 tanggal_peminjaman TEXT,
                 tenggat_pengembalian TEXT,
                 tanggal_pengembalian TEXT,
                 denda REAL,
                 FOREIGN KEY (id_buku_per_copy) REFERENCES buku_per_copy(id_buku_per_copy)
                 FOREIGN KEY (id_anggota) REFERENCES anggota(id_anggota)
                 FOREIGN KEY (id_penjaga) REFERENCES penjaga_perpustakaan(id_penjaga)
                 )''')

# Buat tabel klasifikasi
cursor.execute('''CREATE TABLE klasifikasi (
                 id_klasifikasi TEXT PRIMARY KEY,
                 klasifikasi TEXT
                 )''')

# Buat tabel pengarang
cursor.execute('''CREATE TABLE pengarang (
                 id_pengarang TEXT PRIMARY KEY,
                 nama_pengarang TEXT,
                 email_pengarang TEXT,
                 tempat_tinggal TEXT)''')

# Buat tabel penerbit
cursor.execute('''CREATE TABLE penerbit (
                 id_penerbit TEXT PRIMARY KEY,
                 nama_penerbit TEXT,
                 telepon INTEGER,
                 email_penerbit TEXT)''')


# Buka file CSV buku
with open('DUMMY DATABASE BUKU.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        cursor.execute("INSERT INTO buku (ID_Buku, judul, pengarang, penerbit, Tahun_terbit, klasifikasi) VALUES (?, ?, ?, ?, ?, ?)",
                       tuple(row))

# Buka file CSV anggota
with open('DUMMY DATABASE ANGGOTA.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 6:  # Check if the row has 6 values
            cursor.execute("INSERT INTO anggota (id_anggota, nama_anggota, nik, ttl, alamat, nomor_telepon) VALUES (?, ?, ?, ?, ?, ?)", row)
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV penjaga perpustakaan
with open('DUMMY DATABASE PENJAGA_PERPUSTAKAAN.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 4:  # Check if the row has 7 values
            cursor.execute("INSERT INTO penjaga_perpustakaan (id_penjaga, nama_penjaga, jadwal_kerja, jam_kerja) VALUES (?, ?, ?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV buku copy
with open('DUMMY DATABASE BUKU_COPY.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 3:  # Check if the row has 7 values
            cursor.execute("INSERT INTO buku_per_copy (id_buku_per_copy, id_buku, tersedia) VALUES (?, ?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV peminjaman
with open('DUMMY DATABASE PEMINJAMAN.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 8:  # Check if the row has 7 values
            cursor.execute("INSERT INTO peminjaman (id_peminjaman, id_buku_per_copy, id_anggota, id_penjaga, tanggal_peminjaman, tenggat_pengembalian, tanggal_pengembalian, denda) VALUES (?, ?, ?, ?, ?, ?, ?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV klasifikasi
with open('DUMMY DATABASE KLASIFIKASI.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 2:  # Check if the row has 7 values
            cursor.execute("INSERT INTO klasifikasi (id_klasifikasi, klasifikasi) VALUES (?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV pengarang
with open('DUMMY DATABASE PENGARANG.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 4:  # Check if the row has 7 values
            cursor.execute("INSERT INTO pengarang (id_pengarang, nama_pengarang, email_pengarang, tempat_tinggal) VALUES (?, ?, ?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Buka file CSV penerbit
with open('DUMMY DATABASE PENERBIT.csv', 'r') as berkas_csv:
    data_csv = csv.reader(berkas_csv, delimiter=',')
    next(data_csv)  # Skip header row
    data = list(data_csv)
    for row in data:
        if len(row) == 4:  # Check if the row has 7 values
            cursor.execute("INSERT INTO penerbit (id_penerbit, nama_penerbit, telepon, email_penerbit) VALUES (?, ?, ?, ?)", (row))
        else:
            print(f"Skipping row: {row} (incorrect number of values)")

# Simpan perubahan ke database
conn.commit()

# Tutup koneksi ke database
conn.close()
```

### Input Pengembalian

```python
from ipywidgets import Layout, widgets, Style
from IPython.display import display
from datetime import datetime, timedelta

# Membuat widget input
id_peminjaman_input = widgets.Text(
    value='',
    description='ID Peminjaman:',
    disabled=False)

id_anggota_output = widgets.Text(
    value='',
    description='ID Anggota:',
    disabled=True)

id_buku_output = widgets.Text(
    value='',
    description='ID Buku:',
    disabled=True)

tanggal_kembali_input = widgets.DatePicker(
    description='Tanggal Pengembalian:',
    disabled=False,
    value=datetime.now().date())

submit_button = widgets.Button(
    description='Masukkan Data',
    disabled=False,
    button_style='success',
    tooltip='Klik untuk memasukkan data pengembalian')

def fetch_peminjaman_data(change):
    with sqlite3.connect('perpustakaan.db') as conn:
        c = conn.cursor()
        try:
            c.execute("SELECT id_anggota, id_buku_per_copy FROM peminjaman WHERE id_peminjaman = ?", (id_peminjaman_input.value,))
            result = c.fetchone()
            if result:
                id_anggota, id_buku = result
                id_anggota_output.value = id_anggota
                id_buku_output.value = id_buku
            else:
                id_anggota_output.value = ''
                id_buku_output.value = ''
        except Exception as e:
            print("Terjadi kesalahan: ", e)

id_peminjaman_input.observe(fetch_peminjaman_data, names='value')




def update_tersedia_value(id_buku):
    # Connect to the SQLite database
    connection = sqlite3.connect('perpustakaan.db')
    cursor = connection.cursor()

    # Execute the SQL UPDATE statement to update the tersedia column
    cursor.execute("UPDATE buku_per_copy SET tersedia = ? WHERE id_buku_per_copy = ?;", (1,id_buku_output.value,))

    # Commit the changes and close the connection
    connection.commit()
    connection.close()


def insert_data(b):
    with sqlite3.connect('perpustakaan.db') as conn:
        c = conn.cursor()
        c.execute('PRAGMA foreign_keys = ON')
        try:
            # Mendapatkan data peminjaman berdasarkan id_peminjaman
            c.execute("SELECT tanggal_peminjaman, tenggat_pengembalian FROM peminjaman WHERE id_peminjaman = ?", (id_peminjaman_input.value,))
            result = c.fetchone()
            if result:
                tanggal_peminjaman, tenggat_pengembalian = result
                tanggal_kembali = tanggal_kembali_input.value

                # Menghitung denda
                if tanggal_kembali > datetime.strptime(tenggat_pengembalian, '%Y-%m-%d').date():
                    denda = (tanggal_kembali - datetime.strptime(tenggat_pengembalian, '%Y-%m-%d').date()).days * 5000
                else:
                    denda = 0

                # Memperbarui data peminjaman dengan tanggal_pengembalian dan denda
                c.execute("UPDATE peminjaman SET tanggal_pengembalian = ?, denda = ? WHERE id_peminjaman = ?",
                          (str(tanggal_kembali), denda, id_peminjaman_input.value))
                conn.commit()
                print(f"Data pengembalian berhasil dimasukkan. Denda: Rp{denda}")
            else:
                print("Tidak ada data peminjaman yang ditemukan.")
        except Exception as e:
            print("Terjadi kesalahan: ", e)

submit_button.on_click(insert_data)
submit_button.on_click(update_tersedia_value)
display(id_peminjaman_input, id_anggota_output, id_buku_output, tanggal_kembali_input, submit_button)
```

```python
# Function to fetch IDs from the SQL table
def get_ids_from_db():
    connection = sqlite3.connect('perpustakaan.db')  # Replace with your database file
    cursor = connection.cursor()

    cursor.execute("SELECT id_peminjaman FROM peminjaman;")  # Replace 'id' and 'your_table' with actual column and table names
    ids = cursor.fetchall()

    cursor.close()
    connection.close()

    return [str(id[0]) for id in ids]

# Create a dropdown widget to select an ID
id_dropdown = widgets.Dropdown(
    options=get_ids_from_db(),
    description='ID Peminjaman (yang salah) :',
    disabled=False
)

# Create a button to delete the selected record
delete_button = widgets.Button(
    description='Hapus ID_Peminjaman',
    disabled=False,
    button_style='danger',  # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to delete',
    icon='trash'
)

# Function to handle the deletion
def delete_record(button):
    id_to_delete = id_dropdown.value
    if id_to_delete:
        connection = sqlite3.connect('perpustakaan.db')  # Replace with your database file
        cursor = connection.cursor()

        cursor.execute("DELETE FROM peminjaman WHERE id_peminjaman = ?;", (id_to_delete,))  # Replace 'id' and 'your_table' with actual column and table names
        connection.commit()

        cursor.close()
        connection.close()

        # Update dropdown options
        id_dropdown.options = get_ids_from_db()
        print(f"Record with ID {id_to_delete} deleted successfully.")

# Display the widgets
display(id_dropdown, delete_button)

# Attach the event handler to the button
delete_button.on_click(delete_record)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import sqlite3
from datetime import datetime

# Create search input widget
search_input = widgets.Text(
    value='',
    placeholder='Masukkan kata kunci pencarian',
    description='Cari Buku:',
    style={'description_width': 'initial'},
    disabled=False
)

# Create a button to initiate the search
search_button = widgets.Button(
    description='Cari',
    button_style='primary',
    tooltip='Klik untuk mencari buku berdasarkan kata kunci'
)

# Create an output widget to display search results
output_search_results = widgets.Output()

# Create an output widget to display book availability
output_info_buku = widgets.Output()

# Function to search for books based on the keyword
def search_books(b):
    with output_search_results:
        clear_output()
        keyword = search_input.value

        # Connect to the database
        conn = sqlite3.connect('perpustakaan.db')
        c = conn.cursor()

        # Perform a search query to include the author's name
        c.execute('''
            SELECT b.judul, p.nama_pengarang
            FROM buku b
            JOIN pengarang p ON b.pengarang = p.id_pengarang
            WHERE b.judul LIKE ? OR p.nama_pengarang LIKE ?
        ''', ('%' + keyword + '%', '%' + keyword + '%'))
        results = c.fetchall()

        conn.close()

        if results:
            listjudul = [f"{item[0]} oleh {item[1]}" for item in results]
            id_buku_widget.options = listjudul
            print(f"Ditemukan {len(results)} buku:")
            for judul in listjudul:
                print(f"- {judul}")
        else:
            print("Tidak ditemukan buku dengan kata kunci tersebut.")

# Create a dropdown widget for selecting a book title
id_buku_widget = widgets.Dropdown(
    options=[],
    description='Judul Buku:',
    style={'description_width': 'initial'}
)

# Create a button to check the availability of the selected book
cek_ketersediaan_button = widgets.Button(
    description='Cek Ketersediaan',
    button_style='warning',
    tooltip='Cek ketersediaan buku berdasarkan judul'
)

# Function to check the availability of the selected book
def cek_info_buku(b):
    with output_info_buku:
        clear_output()
        judul_buku = id_buku_widget.value

        if not judul_buku:
            print("Silakan pilih judul buku terlebih dahulu.")
            return

        # Extract the actual book title from the selected option
        judul_buku_actual = judul_buku.split(' oleh ')[0]

        conn = sqlite3.connect('perpustakaan.db')
        c = conn.cursor()

        # Query to get book information and availability including the author's name
        c.execute('''
            SELECT b.judul, p.nama_pengarang, b.id_buku, COUNT(bpc.id_buku_per_copy) AS total_copy,
                   COUNT(CASE WHEN bpc.tersedia = 1 THEN 1 END) AS banyak_tersedia,
                   GROUP_CONCAT(CASE WHEN bpc.tersedia = 1 THEN bpc.id_buku_per_copy END) AS available_copies
            FROM buku b
            JOIN pengarang p ON b.pengarang = p.id_pengarang
            LEFT JOIN buku_per_copy bpc ON b.id_buku = bpc.id_buku
            WHERE b.judul = ?
            GROUP BY b.judul, p.nama_pengarang, b.id_buku
        ''', (judul_buku_actual,))
        info_buku = c.fetchone()

        conn.close()

        if info_buku:
            judul, pengarang, id_buku, total_copy, banyak_tersedia, available_copies = info_buku
            available_copies = available_copies if available_copies else "Tidak ada salinan yang tersedia"
            print(f"Judul: {judul}\nPenulis: {pengarang}\nID Buku: {id_buku}\nTotal Salinan: {total_copy}\nTersedia: {banyak_tersedia}\nID Buku Per Copy Tersedia: {available_copies}")
        else:
            print("Buku dengan judul tersebut tidak ditemukan.")

# Bind the search button to the search function
search_button.on_click(search_books)

# Bind the check availability button to the check function
cek_ketersediaan_button.on_click(cek_info_buku)

# Display the search input, search button, dropdown for selecting book title, check availability button, and output widgets
display(search_input, search_button, output_search_results, id_buku_widget, cek_ketersediaan_button, output_info_buku)
```
