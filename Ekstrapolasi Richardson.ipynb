```python
#mengimport modul yang diperlukan untuk perhitungan dan pembuatan tabel
import sympy 
from numpy import sin,cos,tan,arctan,log,sqrt,exp,pi,e
import numpy as np
from tabulate import tabulate

#membuat fungsi TUGASMETNUM3 yang mengemas keseluruhan program
def TUGASMETNUM3():
    #membuat print-an judul 
    ulang = True #mengassign var ulang = True agar while loop berjalan
    while ulang == True: #membuat while loop selama var ulang bernilai True
        print('<<< Tugas3Metnum >>>')
        print()
        print('Program ini akan mengaproksimasi nilai turunan f(x) dengan metode diferensiasi numerik dan ekstrapolasi Richardson')
    
        NPM = int(input('Masukkan NPM Anda: ')) #inputan memasukkan NPM disimpan ke dalam variabel NPM
        k = 1+(NPM % 4) #menghitung nilai k dari inputan npm
        print('Sehingga k =',k)

        formula = input('Masukkan rumus f(x), boleh bergantung dengan k: f(x) = ') #menyimpan inputan fungsi ke dalam formula

        x = sympy.symbols('x') #mengassign huruf x menjadi variabel

        def f(x,k): #membuat fungsi untuk formula dengan dua parameter yaitu x dan k karena fungsi yang dimasukkan masih mengandung k
            return eval(formula)
        #membuat inputan nilai x(kemudian disimpan ke var x0), h1(inputan untuk h differensiasi numerik), h2(inputan untuk h ekstrapolasi richardson) 
        x0 = eval(input('Masukkan nilai x: '))
        h1 = eval(input('Masukkan nilai h untuk diferensiasi numerik: '))
        h2 = eval(input('Masukkan nilai h untuk ekstrapolasi Richardson: '))
        
        #MEMBUAT FUNGSI DIFFERENSIASI NUMERIK
        def FBDiff(x,h): #Forward/Backward-Difference
            return (f(x+h,k)-f(x,k))/h

        def TPEP(x,h): # Three-Point End Point
            return (1/(2*h)) * (-3*f(x,k) + 4*f(x+h,k) - f(x+2*h,k))

        def TPMP(x,h): # Three-Point Mid Point
            return (1/(2*h)) * (f(x+h,k) - f(x-h,k))

        def FPEP(x,h): # Five-Point End Point
            return (1/(12*h)) * (-25*f(x,k) + 48*f(x+h,k) - 36*f(x+2*h,k) + 16*f(x+3*h,k) - 3*f(x+4*h,k))

        def FPMP(x,h): # Five-Point Mid Point
            return (1/(12*h)) * (f(x-2*h,k) - 8*f(x-h,k) + 8*f(x+h,k) - f(x+2*h,k))

        def SDMP(x,h): # Second Derivative Mid Point
            return (1/(h**2)) * (f(x-h,k) - 2*f(x,k) + f(x+h,k))

        print() #menjaga jarak dari print sebelumnya
        #MEMBUAT PRINT-AN MASING-MASING METODE
        print("Turunan pertama dari f di x0 = {0} adalah : ".format(x0))
        print("f'({0}) = {1} (Forward/Backward-Difference)".format(x0,FBDiff(x0,h1))) #memunculkan hasil metode FBDiff
        print("f'({0}) = {1} (Three Point Endpoint)".format(x0,TPEP(x0,h1))) #memunculkan hasil metode TPEP
        print("f'({0}) = {1} (Three Point Midpoint)".format(x0,TPMP(x0,h1))) #memunculkan hasil metode TPMP
        print("f'({0}) = {1} (Five Point Endpoint)".format(x0,FPEP(x0,h1))) #memunculkan hasil metode FPEP
        print("f'({0}) = {1} (Five Point Midpoint)".format(x0,FPMP(x0,h1))) #memunculkan hasil metode FPMP
        print() #menjaga jarak dari print sebelumnya
        print("Turunan kedua dari f di x0 = {0} adalah : ".format(x0)) 
        print("f''({0}) = {1} (Second Derivative Midpoint)".format(x0,SDMP(x0,h1))) #memunculkan hasil turunan kedua dari metode SDMP
        print() #menjaga jarak dari print sebelumnya
        
        #membuat kondisi untuk pemberitahuan (k=1, k=2, k=3 dan k=4) pada metode apa ekstrapolasi richardson diterapkan
        if k == 1:
            print('Karena k = 1, ekstrapolasi Richardson akan diterapkan pada metode Three-Point Endpoint (TPEP)')
        if k == 2:
            print('Karena k = 2, ekstrapolasi Richardson akan diterapkan pada metode Three-Point Midpoint (TPMP)')
        if k == 3:
            print('Karena k = 3, ekstrapolasi Richardson akan diterapkan pada metode Five-Point Endpoint (FPEP)')
        if k == 4:
            print('Karena k = 4, ekstrapolasi Richardson akan diterapkan pada metode Five-Point Midpoint (FPMP)')

        # fungsi untuk menambahkan baris baru pada tabel yang sudah ada
        def TambahkanRichardsonO2j(tabel_lama, N1baru):
        # Duplikasi tabel lama menjadi tabel baru (karena akan dimodifikasi)
            tabel_baru = []
            for baris in tabel_lama:
                calon_baris = []
                for nilai in baris:
                    calon_baris.append(nilai)
                tabel_baru.append(calon_baris)
            # Tambahkan kolom kosong pada baris-baris yang sudah ada
            for i in range(len(tabel_baru)):
                tabel_baru[i].append("")
            # Membuat baris baru...
            baris_baru = [N1baru]
            # ... dengan kolom sebanyak len(tabel_lama)+1:
            for i in range(len(tabel_lama)):
                baris_baru.append("")
            # meskipun saat ini kosong, setelah ini akan diisi sesuai rumus.
            # Tambahkan dulu ke tabel (sebagai baris paling bawah terbaru):
            tabel_baru.append(baris_baru)
            # Mengisi baris paling bawah
            k = len(baris_baru) # banyaknya titik termasuk titik baru
            i = k-1 # baris baru adalah baris ke-k, dengan indeks (k-1)
            for j in range(1, k): # untuk kolom N2 (indeks 1), N3 (indeks 2), ...
            # N_{j-1} (h) yaitu ke atas satu langkah, ke kiri satu langkah
                Nj1h = tabel_baru[i-1][j-1]
                 # N_{j-1} (h/2) yaitu ke kiri satu langkah
                Nj1h2 = tabel_baru[i][j-1]
                # Pada rumus, "j" yang dimaksud bukanlah indeks, tapi kolom ke-j,
                # sehingga indeks 1 adalah kolom ke-2, indeks 2 adalah kolom ke-3, dst
                j_kolom = j+1
                # nilai baru, N_j (h), menggunakan rumus rekursif
                tabel_baru[i][j] = Nj1h2 + (Nj1h2 - Nj1h)/(4**(j_kolom-1) - 1)
                # Tabel sudah jadi
            return tabel_baru

    # Kode utama untuk Ekstrapolasi Richardson O(h^2j)
        def EkstrapolasiRichardsonO2j(list_N1):
            # Awal membuat tabel
            tabel_mentah = [
                [list_N1[0]]
            ]

            # mula-mula, hanya ada satu nilai yaitu N1 (h),
            # sehingga hanya ada satu baris dan satu kolom
        
            # banyaknya baris/kolom untuk tabel yang akan dibuat
            t = len(list_N1)
            # lakukan TambahkanRichardsonO2j untuk tiap titik berikutnya
            for i in range(1, t):
                tabel_mentah = TambahkanRichardsonO2j(tabel_mentah, list_N1[i])
            # Mengolah tabel menggunakan tabulate
            list_header = [] #membuat list_header untuk menyimpan judul-judul kolom
            #disini kita bagi jadi dua kondisi, saat k = 1 atau 2 dan saat k=3 atau 4 
            #Tabel Richardson untuk metode three point
            if k == 1 or k == 2: 
                for i in range(t): 
                    list_header.append("O(h^{0})".format(2*(i+1))) #membuat format 2*(i+1) agar kolom pertama dimulai dari O(h^2) utk k=1 atau 2
                tabel_olahan = tabulate(tabel_mentah, headers=list_header,
                                    tablefmt="orgtbl")
            #Tabel Richardson untuk metode five point
            if k == 3 or k == 4:
                for i in range(t):
                    list_header.append("O(h^{0})".format(2*(i+2))) #membuat format 2*(i+2) agar kolom pertama dimulai dari O(h^4) utk k=3 atau 4
                tabel_olahan = tabulate(tabel_mentah, headers=list_header,
                                     tablefmt="orgtbl")
            print("Tabel Ekstrapolasi Richardson untuk O(h^2j)")
            print(tabel_olahan)

        #Code untuk menjalankan ekstrapolasi richardson sebagaimana diminta nilai N1 nya berdasarkan metode differensiasi numerik di atas dan diberi syarat k
        list_N1 = [] #buat tampungan list_N1 untuk nilai-nilai N1 dari masing-masing metode yang memenuhi syarat k
        for i in range(4): #membuat for loop dalam range 4 karna yang diminta adalah N1(h), N1(h/2), N1(h/4) dan N1(h/8). Jadi selalu ada 4 baris di kolom pertama
            pembagi = 2**i #membuat var pembagi sebagai 2 pangkat i agar pembagi untuk h nya jalan (dari h/1 -> h/2 -> h/4 dst.)
            h3 = h2/pembagi #membuat var h3 sebagai h2 dibagi pembagi agar nilai h2 yang tadi diinput untuk richardson bisa dibagi sesuai var pembagi (h2/1 -> h2/2 -> h2/4 dst.)
            
            #membuat syarat k, ketika k=1 nilai N1 nya akan diambil dari metode TPEP, ketika k=2 nilai N1 nya akan diambil dari metode TPMP
            #ketika k=3 nilai N1 nya akan diambil dari metode FPEP dan ketika k=4 nilai N1 akan diambil dari metode FPMP
            if k == 1:
                N1baru = TPEP(x0,h3)
            elif k == 2:
                N1baru = TPMP(x0,h3)
            elif k == 3:
                N1baru = FPEP(x0,h3)
            else:
                N1baru = FPMP(x0,h3)

            list_N1.append(N1baru) #mengappend semua nilai dari N1 (sesuai kondisi di atas) ke dalam list_N1

        print() #menjaga jarak dengan print sebelumnya
        EkstrapolasiRichardsonO2j(list_N1) #memanggil fungsi EkstrapolasiRichardson02j dengan parameter list_N1 untuk menjalankan metode richardsonnya
        print() #menjaga jarak dengan print sebelumnya
        #membuat kode agar program dapat menerima input berkali-kali tanpa harus run ulang
        tanya = str(input("Apa mau ulang? (isi Y/N): ")) #membuat variabel tanya untuk menyimpan inputan (Pilihan input hanya Y/N!)
        if tanya == "N": #membuat kondisi ketika tanya mengandung string N
            print("Yeay selesai")
            ulang = False #maka program berhenti
        
TUGASMETNUM3() #memanggil fungsi TUGASMETNUM3 yang mengemas keseluruhan program
```
